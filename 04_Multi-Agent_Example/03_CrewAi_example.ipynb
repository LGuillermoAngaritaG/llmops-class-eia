```python
from crewai import Agent, Crew, Process, Task
from crewai_tools import SerperDevTool

# Remove the @CrewBase decorator and related imports since they're causing issues in notebook
class LatestAiDevelopmentCrew():
    """LatestAiDevelopment crew"""
    
    def researcher(self) -> Agent:
        """
        Creates a researcher agent.
        
        :return: Agent configured for AI research
        :rtype: Agent
        """
        return Agent(
            name='AI Research Analyst',
            role='Research Analyst specialized in AI developments',
            goal='Find and analyze the latest developments in AI',
            backstory='Expert research analyst with deep knowledge of AI industry',
            verbose=True,
            tools=[SerperDevTool()],
            llm=ChatGroq(model_name="mixtral-8x7b-32768")  # Add this line
        )
    
    def reporting_analyst(self) -> Agent:
        """
        Creates a reporting analyst agent.
        
        :return: Agent configured for report writing
        :rtype: Agent
        """
        return Agent(
            name='Report Writer',
            role='Technical writer specialized in AI reports',
            goal='Create clear and concise reports about AI developments',
            backstory='Experienced technical writer with expertise in AI documentation',
            verbose=True,
            llm=ChatGroq(model_name="mixtral-8x7b-32768")  # Add this line
        ) verbose=True
        )

    def research_task(self) -> Task:
        """
        Creates a research task.
        
        :return: Task configured for research
        :rtype: Task
        """
        return Task(
            description='Research the latest developments in {topic}',
            agent=self.researcher(),
            expected_output='Comprehensive research findings about the specified topic'
        )

    def reporting_task(self) -> Task:
        """
        Creates a reporting task.
        
        :return: Task configured for report writing
        :rtype: Task
        """
        return Task(
            description='Create a detailed report based on the research findings',
            agent=self.reporting_analyst(),
            expected_output='Well-structured markdown report'
        )

    def get_crew(self) -> Crew:
        """
        Creates the LatestAiDevelopment crew.
        
        :return: Configured Crew instance
        :rtype: Crew
        """
        return Crew(
            agents=[self.researcher(), self.reporting_analyst()],
            tasks=[self.research_task(), self.reporting_task()],
            process=Process.sequential,
            verbose=True
        )
```

```python
crew = LatestAiDevelopmentCrew().get_crew()
result = crew.kickoff(inputs={'topic': 'AI Agents'})
```
